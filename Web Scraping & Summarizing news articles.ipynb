```
#Import the libraries
import nltk
from newspaper import Article
```

```
#Get the article
url = 'https://www.washingtonpost.com/technology/2019/07/17/you-downloaded-faceapp-heres-what-youve-just-done-your-privacy/?noredirect=on&utm_term=.1938589d078f'
article = Article(url)
```

```
article.download()
article.parse()
nltk.download("punkt")
article.nlp()
```

```
#Get the authors
article.authors
```

```
#Get the published dates
article.publish_date
```

```
#Get the top image of article
article.top_image 
```

```
#Get the article's text
print(article.text)
```

```
#Get summary
print(article.summary)
```
